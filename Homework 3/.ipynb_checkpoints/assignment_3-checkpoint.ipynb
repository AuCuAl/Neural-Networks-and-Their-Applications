# 宋诗生成任务报告：基于 Transformer 架构的实现

## 1. 实验背景
本项目旨在利用深度学习技术自动生成宋诗。传统的古诗生成通常采用 RNN（如 LSTM）架构，但为了提升模型对长距离依赖的建模能力并利用并行计算优势，本实验采用了 **Transformer** 架构。

## 2. 任务要求
- **架构要求**：弃用 RNN，改用 Transformer。
- **开篇词**：以“明月”作为总起词生成古诗。
- **格式要求**：生成固定格式（如七言绝句）。
- **可视化**：展示 Training Loss 的收敛过程。

## 3. 数据预处理
数据集包含多个 JSON 文件。为了生成格式规整的七言绝句，我们对数据进行了以下处理：
1. **过滤**：仅保留包含 4 句、每句 7 个字（加上标点共 32 字符）的诗。
2. **分词**：采用字符级分词。
3. **特殊字符**：引入 `<BOS>` (开始), `<EOS>` (结束), `<PAD>` (填充)。

```python
import pandas
import json
import torch
import torch.nn as nn
import math
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader

# 加载并过滤数据集
def load_data(file_paths):
    all_poems = []
    for file_path in file_paths:
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            for item in data:
                paragraphs = item.get("paragraphs", [])
                poem_text = "".join(paragraphs)
                # 过滤出七言绝句 (28字 + 4标点 = 32字符)
                if len(poem_text) == 32 and len(paragraphs) == 2:
                    all_poems.append(poem_text)
        except FileNotFoundError:
            continue
    return all_poems

# 设置文件列表
file_list = ['poet.song.40000.json', 'poet.song.41000.json', 'poet.song.42000.json', 'poet.song.43000.json']
poems = load_data(file_list)
print(f"提取到符合条件的七言绝句数量: {len(poems)}")

# 构建词表
chars = set("".join(poems))
chars.add("<BOS>")
chars.add("<EOS>")
chars.add("<PAD>")
char2idx = {ch: i for i, ch in enumerate(chars)}
idx2char = {i: ch for i, ch in enumerate(chars)}
vocab_size = len(chars)

class PoetryDataset(Dataset):
    def __init__(self, poems, char2idx):
        self.data = []
        for poem in poems:
            encoded = [char2idx["<BOS>"]] + [char2idx[ch] for ch in poem] + [char2idx["<EOS>"]]
            self.data.append(torch.tensor(encoded, dtype=torch.long))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = self.data[idx][:-1]
        y = self.data[idx][1:]
        return x, y

dataset = PoetryDataset(poems, char2idx)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

## 4. Transformer 模型架构
我们实现了一个基于 Decoder-only 结构的 Transformer 模型，关键组件包括：
- **Positional Encoding**：引入位置信息。
- **Causal Mask**：下三角掩码，确保自回归生成。
- **Transformer Encoder Layers**：提取特征。

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

class PoetryTransformer(nn.Module):
    def __init__(self, vocab_size, d_model=256, nhead=8, num_layers=4):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.pos_encoder = PositionalEncoding(d_model)
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc_out = nn.Linear(d_model, vocab_size)
        self.d_model = d_model

    def generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask

    def forward(self, x):
        mask = self.generate_square_subsequent_mask(x.size(1)).to(x.device)
        x = self.embedding(x) * math.sqrt(self.d_model)
        x = self.pos_encoder(x)
        x = self.transformer(x, mask=mask)
        return self.fc_out(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = PoetryTransformer(vocab_size).to(device)
```

## 5. 模型训练
使用交叉熵损失函数和 Adam 优化器。

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
epochs = 50
epoch_losses = []

model.train()
for epoch in range(epochs):
    total_loss = 0
    for batch_idx, (x, y) in enumerate(dataloader):
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        output = model(x)
        loss = criterion(output.reshape(-1, vocab_size), y.reshape(-1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
            
    avg_loss = total_loss / len(dataloader)
    epoch_losses.append(avg_loss)
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f'Epoch {epoch+1} Average Loss: {avg_loss:.4f} ====')
```

## 6. 古诗生成与可视化
生成以“明月”开头的诗句并绘制 Loss 曲线。

```python
def generate_poem(model, start_text="明月", max_length=32):
    model.eval()
    input_seq = [char2idx["<BOS>"]] + [char2idx[ch] for ch in start_text]
    input_tensor = torch.tensor(input_seq, dtype=torch.long).unsqueeze(0).to(device)
    
    with torch.no_grad():
        for _ in range(max_length - len(start_text)):
            output = model(input_tensor)
            next_token = torch.argmax(output[0, -1, :]).item()
            if next_token == char2idx["<EOS>"]: break
            input_tensor = torch.cat([input_tensor, torch.tensor([[next_token]]).to(device)], dim=1)
            
    return "".join([idx2char[idx] for idx in input_tensor.squeeze().tolist()[1:]])

print("【生成演示】:")
print(generate_poem(model, start_text="明月"))

plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs + 1), epoch_losses, marker='o', color='blue', label='Train Loss')
plt.title('Training Loss Curve (Transformer)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.savefig('Figures/Figure_1.png', dpi=300, bbox_inches='tight')
plt.show()
```
